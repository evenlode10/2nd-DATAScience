# Matplotlib Basics

Let's start by assigning an array of values for the x axes, and for the y axes

```python
x = np.arange(1, 10, 1)
y = x**2

print(x)
print(y)
```

# There are two different ways to draw plots (lines)

## First, the functional appproach:

```python
# Functional approach 

plt.plot(x, y) 

plt.show()
```

```python
# Functional approach 

plt.plot(x, y, 'r')   # r for red - from matlab

plt.xlabel('x axes label')
plt.xlabel('y axes label')
plt.title('Graph') 

plt.show()
```

```python
# Functional approach 

plt.subplot(1,2,1)
plt.plot(x, y, 'r')   # r for red - from matlab

plt.subplot(1,2,2)
plt.plot(y, x, 'b')# b for red - from matlab
```

# Object-Oriented Approach

- fig / figure is like a blank canvas, that we draw on
- the axes have to be drawn a little way in from the left and bottom edge - and then stop a little way before the edge of the canvas

```python
# Object-Oriented Approach

fig = plt.figure() #like a blank canvas

#Now plot the axes on the canvas - draw the lines - 10% in from the left edge 


                    #left, bottom, width, height - in from the edges of the canvas
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])   #takes four arguments
```

```python
# Object-Oriented Approach

fig = plt.figure() #like a blank canvas

                    #left, bottom, width, height
axes = fig.add_axes([0.1, 0.1, 1, 1])   #takes four arguments
axes.plot(x,y)
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('The title')
```

# Multiple plots on axes

```python
fig = plt.figure(figsize=(4,4))

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**2)
ax.plot(x,x**3)

```

# Using a legend to distinguish between plots

```python
fig = plt.figure(figsize=(4,4))

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**2, label = "X Squared")
ax.plot(x,x**3, label = "X Cubed")

ax.legend() # requires labels to be given to the plots
```

# Exploratory data analysis

```python
import matplotlib.pyplot as plt
import numpy as np

N = 4

survived = (70, 80, 20, 300 )
perished = (20, 30, 319, 100)


ind = np.arange(N)    # the x locations for the groups
width = 0.35       # the width of the bars: can also be len(x) sequence

fig, ax = plt.subplots()

p1 = ax.bar(ind, survived, width, label='Survived')

p2 = ax.bar(ind, perished, width,
            bottom=survived, label='Perished')

ax.axhline(0, color='grey', linewidth=1)
ax.set_ylabel('Count')
ax.set_title('Survival by group and gender')

ax.set_xticks(ind, labels=['Boys', 'Girls', 'Men', 'Women'])
ax.legend()

# Label with label_type 'center' instead of the default 'edge'
ax.bar_label(p1, label_type='center')
ax.bar_label(p2, label_type='center')
ax.bar_label(p2)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery')

# make data:
np.random.seed(3)

x = 0.5 + np.arange(8)
y = np.random.uniform(2, 7, len(x))

# plot
fig, ax = plt.subplots()

ax.bar(x, y, width=1, edgecolor="white", linewidth=0.7)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))



plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery')

# make the data
np.random.seed(3)
x = 4 + np.random.normal(0, 2, 24)
y = 4 + np.random.normal(0, 2, len(x))

# size and color:
sizes = np.random.uniform(15, 80, len(x))
colors = np.random.uniform(15, 80, len(x))

# plot
fig, ax = plt.subplots()

ax.scatter(x, y, s=sizes, c=colors, vmin=0, vmax=100)


ax.set(xlim=(0, 8), xticks = np.arange(1,8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```
